```python
dict_list = []
for i in range(1,74):
    tmp_dict = {
        "name": f"f_{i}-list",
        "value_count": {"min":2, "max":101},
        "type": "INT",
        "int_domain":{
            "name":f"f_{i}",
            "min":0,
            "max":0,
            "is_categorical": True
        },
        "annotation": {
            "tag1": "categorical",
            "tag2": "list"
          }
    }
    dict_list.append(tmp_dict)
```

```python
def get_min(x):
    min_v = np.array(x).min()
    return min_v
def get_max(x):
    max_v = np.array(x).max()
    return max_v
```

```python
for i in range(1,74):
    data["min"]= data[f"f_{i}-list"].apply(get_min)
    data["max"]= data[f"f_{i}-list"].apply(get_max)
    min_v = data["min"].min()
    max_v = data["max"].max()
    print(i, min_v, max_v)
    dict_list[i-1]["int_domain"]["min"] = min_v
    dict_list[i-1]["int_domain"]["max"] = max_v
```
