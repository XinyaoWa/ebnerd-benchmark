```python
def get_max(feature):
    max_out = 0
    for l in feature:
        max_tmp = np.array(l).max()
        if max_tmp > max_out:
            max_out = max_tmp
    return max_out

def get_min(feature):
    min_out = 0
    for l in feature:
        min_tmp = np.array(l).min()
        if min_tmp < min_out:
            min_out = min_tmp
    return min_out
```

```python
test["feature_max"] = test["feature"].apply(get_max)
test["feature_max"].max()
```

```python
test["feature_min"] = test["feature"].apply(get_min)
test["feature_min"].min()
```
