```python
import numpy as np
def get_min(x):
    min_v = np.array(x).min()
    return min_v
def get_max(x):
    max_v = np.array(x).max()
    return max_v
```

```python
df["min"] = df["article_id_fixed"].apply(get_min)
```

```python
df["min_inview"] = df["article_ids_inview"].apply(get_min)
df["max_inview"] = df["article_ids_inview"].apply(get_max)
```

```python
df["max"] = df["article_id_fixed"].apply(get_max)
```
